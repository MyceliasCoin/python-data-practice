```python
#---Step 0: Import necessary libraries and render plots inline---

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

np.random.seed(0)

plt.style.use('fivethirtyeight')
```

```python
# Based on stackoverflow answer to groupby (https://stackoverflow.com/questions/23377108/pandas-percentage-of-total-with-groupby?fbclid=IwAR0YY192JMzHYGFhfGyS5WlsaplN9tKQW3kikFNQtOvXSioaL72ScBLL2ns)

df = pd.DataFrame({'state': ['CA', 'WA', 'CO', 'AZ'] * 3,
                   'office_id': list(range(1, 7)) * 2,
                   'sales': [np.random.randint(100000, 999999)
                             for _ in range(12)]})

df.sort_values(["state", "office_id"])
```

```python
state_office = df.groupby(['state', 'office_id']).agg({'sales': 'sum'})

state_office
```

```python
# Groupby state_office and divide by sum (level=0 means you group by first level of index)
state_pcts = state_office.groupby(level=0).apply(lambda x: ((x / x.sum()) * 100))

state_pcts
```
